# **1. Money Change**

**Problem Introduction**

In this problem, you will design and implement an elementary greedy algorithm
used by cashiers all over the world millions of times per day.
Problem Description

Task. The goal in this problem is to find the minimum number of coins needed to change the input value
(an integer) into coins with denominations 1, 5, and 10.

**Input Format.** 

The input consists of a single integer 𝑚.
Constraints. 1 ≤ 𝑚 ≤ 103


**Output Format.**

Output the minimum number of coins with denominations 1, 5, 10 that changes 𝑚.

```python
def GreedyChange(m):
  coins = [10,5,1]
  i = 0
  count = 0
  while m > 0:
    if m - coins[i] >= 0:
      m -= coins[i]
      count += 1
    else: 
      i += 1
  return count

m = int(input())
print(GreedyChange(m))
```

# **2. Maximum Value of The Loot**

```python
def OrderByRatio(values, weight):
  n = len(values)
   
  # Traverse through all array elements
  for i in range(n):
      swapped = False
      # Last i elements are already
      #  in place
      for j in range(0, n-i-1):
  
          # traverse the array from 0 to
          # n-i-1. Swap if the element 
          # found is greater than the
          # next element
          if values[j]/weight[j] > values[j+1]/weight[j+1] :
              values[j], values[j+1] = values[j+1], values[j]
              weight[j], weight[j+1] = weight[j+1], weight[j]
              swapped = True

      # IF no two elements were swapped
      # by inner loop, then break
      if swapped == False:
            break

def MaximumLootGreedy(values, weights, W, n):
  loot = 0
  i = n-1
  OrderByRatio(values, weights)
  while W > 0 and i >= 0:
    finalWeight = min(weights[i],W)
    W -= finalWeight
    loot += (values[i]/weights[i])*finalWeight
    i -=1    
  return loot

values = []
weights = []
n, W = tuple(map(int, input().split()))
for _ in range(n):
  vi, wi = tuple(map(int, input().split()))
  values.append(vi)
  weights.append(wi)

print(MaximumLootGreedy(values, weights, W, n))
```

# **3 Car Fueling**

```python
def GreedyJourney(x,n,tankCapacity,distance):
  numRefills = 0
  currentRefill = 0
  # We include in the stop's array the initial point, and the last point to make it easier
  x[:0] = [0]
  x[len(x):] = [distance]
  # Initial point = x0 <= x1 <= x2 <= ... <= xn <= xn+1 = distance
  while currentRefill <= n:
    lastRefill = currentRefill

    while currentRefill <= n and x[currentRefill+1] - x[lastRefill] <= tankCapacity:
      currentRefill += 1
    if currentRefill == lastRefill:
      return -1
    if currentRefill <= n:
      numRefills += 1
  
  return numRefills
      


d = int(input())
m = int(input())
n = int(input())
stops = list(map(int, input().split()))
print(GreedyJourney(stops,n,m,d))
```

# **4 Max Ad Revenue**

```python
def MaxRevenue(ads, slots):
    ads.sort()
    slots.sort()
    value = 0
    while len(ads) > 0:
        value += ads[0]*slots[0]
        ads = ads[1:]
        slots = slots[1:]
    return value



if __name__ == '__main__':
    N = int(input())
    ads = list(map(int, input().split()))
    slots = list(map(int, input().split()))
    print(MaxRevenue(ads,slots))
```

# **5 Collecting Signatures**

```python
def IsIn(p1,p2,x):
    if x >= p1 and x <= p2:
        return True
    return False


def MinVisits(N,T):
    T.sort(key = lambda T:T[1])
    points = []
    index = 0
    while index < N:
        NewPoint = T[index][1]
        points.append(str(NewPoint))
        while index < N and IsIn(T[index][0],T[index][1],NewPoint) :
            index += 1
    print(len(points))
    print(" ".join(points))
    return


if __name__ == '__main__':
    N = int(input())
    T = []
    for i in range(N):
        a, b = list(map(int, input().split()))
        T.append([a,b])

    MinVisits(N,T)
```

# **6 Maximum number of Prizes**

```python
def MaximumPrizes(N):
  """
  Here the greedy choice is to select always the smallest number possible, so that we are assigning that number of candies to the last contestant
  that way we are always using as small numbers as possible, which will allow us to use more numbers at the end. If thw numbers we used does not 
  reach the entire number of prizes, We add the difference in the last element
  """
  prizes = []
  i = 1
  count = 0
  totalSum = 0
  aux = N
  while N > 0 and N-i >= 0:
    prizes += [str(i)]
    N -= i
    totalSum += i
    i += 1
    count += 1
  if totalSum < aux:
    prizes[len(prizes) - 1] = str(int(prizes[len(prizes) - 1]) + (aux-totalSum))
    
  return count, prizes





N = int(input())
count, prizes = MaximumPrizes(N)
print(count)
print(' '.join(prizes))
```
